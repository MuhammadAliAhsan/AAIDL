```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

from keras.models import Sequential
from keras.layers import Dense

from sklearn import metrics   

from tensorflow.keras.layers import Dropout
import tensorflow as tf
```

```python
traindata = pd.read_csv('/content/train.csv')
traindata
```

```python
sns.countplot(x=traindata['Class'])
```

```python
# extract labels
output = traindata['Class']
output = pd.Series(output)
print(output.value_counts())

# remove unnecessary columns
input = traindata.drop(['Id', 'Class'], axis=1)
print(input.info())
input.head()
```

```python
# split data to train and validation sets
input_train, input_val, output_train, output_val = train_test_split(input, output, test_size=0.3, random_state=100)
print(f'training data set size: {len(input_train)}')
print(f'validation data set size: {len(input_val)}')
```

```python
# define the keras model
model= Sequential()
model.add(Dense(units=70, input_dim=8, activation='relu'))
model.add(Dense(units=35, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
# compile the keras model
import tensorflow as tf
# create an optimiser object and set the learning rate
optmzr = tf.keras.optimizers.Adam(learning_rate=0.01)
```

```python
# Compile the model with the optimiser object
model.compile(loss='binary_crossentropy', optimizer=optmzr, metrics=['accuracy'])
```

```python
# train model
history = model.fit(input_train, output_train, batch_size=10, epochs=60, validation_data=(input_val, output_val ))
```

```python
# summarize the history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Accuracy Of Model')
plt.ylabel('Accuracy')
plt.xlabel('Epoch Value')
plt.legend(['train','test'], loc='upper left')
plt.show()

# summarize the history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch Value')
plt.legend(['train','test'], loc='upper left')
plt.show()
```

```python
# Getting Model Prediction
output_pred = model.predict(input_val)
print(output_pred[:10])
```

```python
# Prediction Conversion
output_pred_categorical1 = [1 if pred > 0.5 else 0 for pred in output_pred]
print(output_pred_categorical1[:10])
```

```python
# Accuracy Measuring
accuracy = metrics.accuracy_score(output_val, output_pred_categorical1)
print(f'Accuracy: {accuracy}')
```
